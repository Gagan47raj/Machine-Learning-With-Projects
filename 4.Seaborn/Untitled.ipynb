```python
import seaborn as sb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv = pd.read_csv("C:/Users/gagan/Desktop/Python/ML/4.Seaborn/StudentsPerformance.csv")
csv
```

```python
sb.lineplot(x="writing score", y="reading score", data = csv)
plt.show()
```

```python
sb.lineplot(x="writing score", y="reading score", data = csv, hue="gender",style="gender", markers=["o",">"])
plt.show()
```

```python
csv1 = csv.head(50)
sb.lineplot(x="writing score", y="reading score", data = csv1, hue="gender",style="gender", markers=["o",">"])
plt.title("Reading and Writing Score on Gender Basis")
plt.show()
```

# Histogram/Distplot

```python
csv['average'] = csv[['math score','reading score','writing score']].mean(axis=1)
```

```python
csv
```

```python
csv
```

```python
sb.histplot(csv['average'])
```

```python
plt.figure(figsize=(16,9))
sb.set()
sb.histplot(csv['average'], label="Average")
plt.title("Histogram of average marks")
plt.legend()
```

# Heatmap
graphical representation of 2D data | particularly useful for showing correlation matrices, categorical data, and numerical values in a grid format | 

```python
array2D = np.linspace(1,5,12).reshape(4,3)
```

```python
array2D
```

```python
sb.heatmap(data=array2D)
```

```python
ign_scores = pd.read_csv("ign_scores.csv")
ign_scores
```

```python
ign_scores = ign_scores.set_index(ign_scores["Platform"])
```

```python
ign_scores
```

```python
ign_scores = ign_scores.drop(columns=["Platform"],axis=1)
```

```python
ign_scores
```

```python
plt.figure(figsize=(16,9))
sb.heatmap(ign_scores)
```

```python
plt.figure(figsize=(16,9))
sb.heatmap(ign_scores, vmin = 0, vmax = 10, annot=True)
```

```python
plt.figure(figsize=(16,9))
ht_map = sb.heatmap(ign_scores, vmin = 0, vmax = 10, annot=True, linewidths=4,linecolor="k")
ht_map.set(title="Heatmap of IGN Games Score", xlabel="Generes", ylabel= "Platform")
```

```python
ign_scores.corr()
```

```python
plt.figure(figsize=(16,9))
ht_map = sb.heatmap(ign_scores.corr(), annot=True, linewidths=4,linecolor="k")
ht_map.tick_params(size=10,color='k',labelsize=10,labelcolor='k')
ht_map.set(title="Heatmap of IGN Games Score", xlabel="Generes", ylabel= "Platform")
plt.show()
```
