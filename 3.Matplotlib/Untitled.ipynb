# Matplotlib

```python
#2D plotting library 
import matplotlib.pyplot as plt
import numpy as np
```

# Linear Graph

```python
sales = [10,20,30,40]
day = [1,2,3,4]

plt.plot(day,sales)
```

```python
sales = [70,30,40,50,80,95,98]
day = [1,2,3,4,5,6,7]

plt.plot(day,sales,marker = 'o')
plt.show()
```

# Bar Graph

```python
sales = [70,30,40,50,80,95,98]
day = [1,2,3,4,5,6,7]

plt.bar(day,sales)
plt.show()
```

```python
sales = [70,30,40,50,80,95,98]
day = [1,2,3,4,5,6,7]
c = ['r','c','b','g','y','k','m']
plt.bar(day,sales, color = c)
plt.show()
```

```python
x = np.array(["Monday","Tuesday","Wednesday","Thurday","Friday","Saturday","Sunday"])
y = np.array([70,30,40,50,80,95,98])

c = ['r','c','b','g','y','k','m']
plt.barh(x,y, color = c)
plt.show()
```

```python
x = np.array(["Monday","Tuesday","Wednesday","Thurday","Friday","Saturday","Sunday"])
y = np.array([70,30,40,50,80,95,98])

c = ['r','c','b','g','y','k','m']
plt.bar(x,y, color = c, width = 0.4)
plt.show()
```

# Scatter Plot

```python
x = np.array(["Monday","Tuesday","Wednesday","Thurday","Friday","Saturday","Sunday"])
y = np.array([70,30,40,50,80,95,98])
c = ['r','c','b','g','y','k','m']

plt.title("Scatter plot of weekly sale")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.scatter(x,y, color = c, s = 80)
plt.show()
```

# Pie Chart

```python
diet = [30,60,10]
name = ["meal","Junk Food", "Fruit and Nuts"]

plt.pie(diet, labels = name)
plt.show()
```
